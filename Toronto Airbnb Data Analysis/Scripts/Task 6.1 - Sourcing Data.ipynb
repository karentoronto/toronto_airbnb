### Checking for Missing Values 

```python
# check for missing values
df.isna().sum()
```

# Plan for missing values 
### drop host_name = column meaningless
### drop 'neightborhood_group' 
###  drop "license" column = doesn't provide meaning
###  for "price": will need to impute, it's missing 19% of the data.



```python
df.drop(columns=['host_name', 'neighbourhood_group'], inplace=True)
```

```python
df.drop(columns=['license'], inplace=True)
```

```python
df.info()
```

```python
# Dealing with "price" column, to impute values
df['price'] = df['price'].fillna(df['price'].median())
```

```python
df.isna().sum()
```

```python
# Dealing with 'reviews_per_month', to impute values
df['reviews_per_month'] = df['reviews_per_month'].fillna(df['reviews_per_month'].median())
```

### Dealing with data types 

```python
# change "last_review" to date format

df['last_review'] = pd.to_datetime(df['last_review'])
```

```python
df.info()
```

### Extract rating from "name" column

```python
# create a new variable to store the extracted rating 
df['rating'] = df['name'].str.extract('(\d+\.\d+)')
```

```python
# checking the new "rating" column
df['rating'].head()
```

```python
# change datatype of 'rating' column
df['rating'] = df['rating'].astype('float16')
```

```python
# checking 
df.isna().sum()
```

### Dealing with missing Data - "rating" and "last_review " columns

```python
# For "rating" column, fill in missing values withe median value of all the ratings
df['rating'] = df['rating'].fillna(df['rating'].median())
```

```python
df.describe()
```

```python
# Note the "last_review" column is missing a significant percentage of data. This means there is no last review date for the rental. 
# Dates can provide valuable information on the rental trends in Toronto. Instead of dropping this information altogether, I'll focus only on a subset
# of the data starting on January 1, 2015 to January 8, 2024. 
df_subset = df[df['last_review'] >= '2015-01-01']
```
